# GRAFICAMOS LA SERIES DE TIEMPO

```python
# iMPORTAMOS LAS LIBRERÍAS PARA GRAFICAR
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# UTILIZAMOS LINEPLOT PARA GRAFICAR
# Ajustamos el tamaño de la figura
plt.figure(figsize=(12, 8))
plt.xlabel('FECHA')
plt.ylabel('PRECIO')
plt.title('PRECIO DE CIERRE ACCIONES DE MICROSOFT')
# Graficamos
sns.lineplot(data=df['Close'])
```

# Personalizamos el gráfico por año y mes con Dateformater

```python
# Personalizamos la escala temporal
# Importamos "dates" de Matplotlib" que nos permiten controlar la forma como mostrarán las marcas de tiempo
import matplotlib.dates as mdates
```

```python
# generamos el gráfico anterior
plt.figure(figsize=(12, 8))
plt.xlabel('FECHA')
plt.ylabel('PRECIO')
plt.title('PRECIO DE CIERRE ACCIONES DE MICROSOFT')
# Personalizamos la escala
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))

# Inclinamos el eje x
plt.xticks(rotation=45)



# Graficamos
sns.lineplot(data=df['Close'])
```

```python
# Genereamos el gráfico anterior
plt.figure(figsize=(12, 8))
plt.xlabel('FECHA')
plt.ylabel('PRECIO')
plt.title('PRECIO DE CIERRE ACCIONES DE MICROSOFT')
# Personalizamos la escala de eje x para mostrar el tiempo en intervalo de cada 5 meses
plt.gca().xaxis.set_major_locator(mdates.MonthLocator(interval=5))

# Personalizamos la escala de eje x para mostrar el año y el nombre del mes
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%b'))

# Inclinamos el eje x
plt.xticks(rotation=45)


# Creamos el gráfico
sns.lineplot(data=df['Close'], alpha = 0.5, palette=["red"])
```

```python
# Genereamos el gráfico anterior
plt.figure(figsize=(12, 8))
plt.xlabel('FECHA')
plt.ylabel('PRECIO')
plt.title('PRECIO DE CIERRE ACCIONES DE MICROSOFT')


# Personalizamos la escala de eje x para mostrar el año y el nombre del mes
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%b-%d'))

# Inclinamos el eje x
plt.xticks(rotation=90)


# Creamos el gráfico
sns.lineplot(data=df['Close'], alpha = 0.5, palette=["red"])
```

# Vamos a resaltar la fecha del 13 de enero al 21 de enero
- Usar "axvspan" para resaltar la porción deseada.

```python
# Genereamos el gráfico anterior
plt.figure(figsize=(12, 8))
plt.xlabel('FECHA')
plt.ylabel('PRECIO')
plt.title('PRECIO DE CIERRE ACCIONES DE MICROSOFT')


# Personalizamos la escala de eje x para mostrar el año y el nombre del mes
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%b-%d'))

# Resaltar el rango de tiempo de interés.
t_inicio = pd.Timestamp("2000-01-13")
t_fin = pd.Timestamp("2000-01-21")
plt.gca().axvspan(t_inicio, t_fin, alpha=0.5, color='red')


# Inclinamos el eje x
plt.xticks(rotation=90)


# Creamos el gráfico
sns.lineplot(data=df['Close'], alpha = 0.5, palette=["red"])
```
